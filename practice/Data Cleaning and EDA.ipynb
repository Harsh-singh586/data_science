```python
gender = pd.get_dummies(data.gender)
```

```python
residence = pd.get_dummies(data.residence)
```

```python
prev_ed = pd.get_dummies(data.prevEducation)
```

```python
concat_data = pd.concat([data, gender, residence, prev_ed], axis=1)
```
