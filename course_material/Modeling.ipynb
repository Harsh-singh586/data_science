## creating table

```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('Data/online_shop.db')
```

```python
transactions = pd.read_sql('SELECT * FROM transactions', conn)
transactions.head()
```

```python
items = pd.read_sql('SELECT * FROM items', conn)
items.head()
```

```python
df = transactions.merge(items, how="left", on="item_id")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.purchase_date = pd.to_datetime(df.purchase_date)
df.price = pd.to_numeric(df.price.str.replace("$", ""))
df.rating = pd.to_numeric(df.rating)
```

```python
df.dtypes
```

```python
df_april_may = df[df.purchase_date.dt.month < 6]
df_april_may.head()
```

```python
df_june = df[df.purchase_date.dt.month == 6]
df_june.head()
```

```python
dog_food_june = df_june[df_june.item_description == 'Dog Food']
```

```python
june_dog_food_purchases = (dog_food_june
  .groupby('customer')['item_id']
  .count()
  .rename('june_dog_food_purchases')
)
```

```python
june_dog_food_purchases
```

```python
april_dog_food_cost = (df_april_may
  .groupby('customer')['price']
  .sum()
  .rename('april_dog_food_cost')
)
```

```python
april_dog_food_cost
```

```python
model_data = pd.concat([june_dog_food_purchases, april_dog_food_cost], axis=1).fillna(0).reset_index()
```

```python
category_dummy = pd.get_dummies(df_april_may.category)
```

```python
model_category = model_data.merge(categories, how="left", on="customer")
```
